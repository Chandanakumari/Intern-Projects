```python
#Exporting required Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import gc
import os
import time
import datetime as dt
import mlxtend.preprocessing
import mlxtend.frequent_patterns
```

```python
#Reading the file
df=pd.read_csv('Merged_data_file.csv')
```

### Note-Performing the market basket analysis only for 2023 data for the period of 6 months as the process time of entire data takes long time

```python
#MBA_df stands for market basket aalysis data frame
#narrowing down the data to the required number of columns
df_2023=df[df['year']==2023]

MBA_df=df_2023[['Order_Number','Lineitem_name']]
```

```python
# the data is already cleaned in the provious EDA notebook
```

```python
# converting the basket item to list of lists
item_list = []

for num in list(set(MBA_df.Order_Number.tolist())):
    # filter data set down to one invoice number
    tmp_df = MBA_df.loc[MBA_df['Order_Number'] == num]
    # extract item descriptions and convert to list
    tmp_items = tmp_df.Lineitem_name.tolist()
    # append list invoice_item_list
    item_list.append(tmp_items)

print(item_list[1:3])
```

```python
# fitting the transaction encoder
online_encoder = mlxtend.preprocessing.TransactionEncoder()
online_encoder_array = online_encoder.fit_transform(item_list)

# converting the array as a dataframe
online_encoder_df = pd.DataFrame(online_encoder_array, columns=online_encoder.columns_)

# Print the results
online_encoder_df
```

# Association Rules with Apriori Algorithm

#### Using Apriori algorithm to analyse the frequent itemsets in this transaction data

```python
#running the algorithm with minimum support of 0.0.1 as default is by 0.5

apriori_model = mlxtend.frequent_patterns.apriori(online_encoder_df, min_support=0.01)
apriori_model
```

```python
#Rinning the algorithm again with column names(item names) replacing the numerical designations
apriori_model_colnames = mlxtend.frequent_patterns.apriori(
    online_encoder_df, 
    min_support=0.01,
    use_colnames=True
)

apriori_model_colnames
```

```python
#adding another column that tells us the size itemsets
apriori_model_colnames['length'] = (
    apriori_model_colnames['itemsets'].apply(lambda x: len(x))
)

apriori_model_colnames
```

```python
#generating a rule with key parameters. Here, we are using confidence as the metric with 0.06
rules = mlxtend.frequent_patterns.association_rules(
    apriori_model_colnames, 
    metric="confidence",
    min_threshold=0.06, 
    support_only=False
)

rules
```

```python
#number of associations
print("Number of Associations: {}".format(rules.shape[0]))
```

```python
#setting metrics to get the most frequent brought  itemset
rules[ (rules['support'] >= 0.02) &
       (rules['confidence'] > 0.06) &
       (rules['lift'] > 6.0) ]
```

### Interpretation 

#### Setting the minimum threshold or metric values for confidence , support and lift totally depends on the business objective and requirements.  I have chosen these values with my best knowledge.

#### With the support values greater than  0.02,confidence greater than  0.06 and lift greated than 6 we find "Sunfeast Caker swiss roll caker chocolate" and "Sunfeast caker trinity cake" to the most frequently brought itemset.

#### The above table gives us the most frequently brought item set with the respective metric value

### The above result explains the best combination products. Taking row 109 as an example, the support value of 0.04 means there is 4% of occurrence of the combination product of antecedent and consequent in the total number of transaction, which means out of 100 baskets,4 combinations are of product A and Product B. Confidence value of 0.52 means , given that antecedent is present in the basket there are 52% of chance that the consequent will be present in the basket. A lift value of 13 means , presence of antecedent has uplifted the chances of presence of consequence by 13 times.
