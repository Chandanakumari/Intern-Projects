3. Diagnosis Numeric- This is encoding of diagnosis where 0 is for B and 1 is for M
4. Radius- Mean of distances from the centre to points on the perimeter cell
5. Texture- The standard deviation of grayscale values
6. perimeter- Perimeter of a cell
7. area- are of a cell
8. smoothness- Local variation in radius lengths
9. compactness- perimeter^2/area-1
10. concavity- severity of concave portions of the contour
11. Symmetry- Symmetry of the cell
12. fractal dimension- coastline appriximation-1

The mean,standard error and 'worst'(mean of the three largest values) of these features were computed for each image ,resulting in 30 features. For example,for radius standard error is radius_se and worst is radius_worst. Similarly for all the fields.

```python
#Creating a function to find the outliers
def find_outliers_IQR(ndf):
    q1=ndf.quantile(0.25)
    q3=ndf.quantile(0.75)
    IQR=q3-q1
    outliers = ndf[((ndf<(q1-1.5*IQR)) | (ndf>(q3+1.5*IQR)))]
    return outliers
```

# Removing outliers

```python
cols = ['radius', 'texture', 'perimeter',
       'area', 'smoothness', 'compactness', 'concavity', 'concave_points',
       'symmetry', 'fractal_dimension', 'radius_se', 'texture_se',
       'perimeter_se', 'area_se', 'smoothness_se', 'compactness_se',
       'concavity_se', 'concave_points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worse', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave_points_worst',
       'symmetry_worst', 'fractal_dimension_worst'] 

Q1 = df[cols].quantile(0.25)
Q3 = df[cols].quantile(0.75)
IQR = Q3 - Q1
# orf stands for outliers removed df
orf = ndf[~((ndf[cols] < (Q1 - 1.5 * IQR)) |(ndf[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
```
